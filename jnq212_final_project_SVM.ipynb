```python
import numpy as np
import pandas as pd
from sklearn import metrics
from sklearn.metrics import confusion_matrix
import seaborn as sns
from sklearn.pipeline import make_pipeline
from sklearn.model_selection import train_test_split  # The standard - train/test to prevent overfitting and choose hyperparameters
from sklearn.metrics import accuracy_score
from sklearn.svm import SVC
import numpy as np
import matplotlib.pyplot as plt 
from matplotlib import pyplot
from sklearn.preprocessing import StandardScaler

train_df = pd.read_csv("fashion-mnist_train.csv")
test_df = pd.read_csv("fashion-mnist_test.csv")

train_data =  np.array(train_df)
test_data =  np.array(test_df)

X_train = np.array(train_data[:,1:])
y_train = np.array(train_data[:,0])
X_test = np.array(test_data[:,1:])
y_test = np.array(test_data[:,0])

# Making the dataset smaller since SVM cannot handle 60,000 data points in a resonable amount of time
X_train_not_used, X_train, y_train_not_used, y_train = train_test_split(X_train, y_train, test_size=5000, random_state=10, shuffle=True)
X_test_not_used, X_test, y_test_not_used, y_test = train_test_split(X_test, y_test, test_size=5000, random_state=10, shuffle=True)
```

```python
# What the examples looks like. Note, the feature  vector was rescaled to be 28x28
for i in range(4):  
  pyplot.imshow(X_train[i].reshape((28,28)), cmap=pyplot.get_cmap('gray'))
  pyplot.show()
```

```python
# Normalizing the features, so each feature is in the range [-1,1].  
# In this datset, each feature has a minimum value of 0 and a maximum value of 255.
X_train = 2*X_train/255 -1
X_test = 2*X_test/255-1
```

```python
# What the dataset looks like after scaling
for i in range(4):  
  pyplot.imshow(X_train[i].reshape((28,28)), cmap=pyplot.get_cmap('gray'))
  pyplot.show()
```

```python
polyA_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="poly", degree=3, coef0=1, C=0.001))
polyB_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="poly", degree=3, coef0=1, C=0.01))
polyC_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="poly", degree=3, coef0=1, C=0.1))
polyD_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="poly", degree=3, coef0=1, C=1))
polyE_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="poly", degree=3, coef0=1, C=10))

linearA_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="linear", C=0.001))
linearB_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="linear", C=0.01))
linearC_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="linear", C=0.1))
linearD_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="linear", C=1))
linearE_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="linear", C=10))


radialA_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", gamma=0.001, C=0.001))
radialB_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", gamma=0.001, C=0.01))
radialC_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", gamma=0.001, C=0.1))
radialD_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", gamma=0.001, C=1))
radialE_kernel_svm_clf = make_pipeline(StandardScaler(), SVC(kernel="rbf", gamma=0.001, C=10))
```

```python
polyA_kernel_svm_clf.fit(X_train, y_train)
polyB_kernel_svm_clf.fit(X_train, y_train)
polyC_kernel_svm_clf.fit(X_train, y_train)
polyD_kernel_svm_clf.fit(X_train, y_train)
polyE_kernel_svm_clf.fit(X_train, y_train)

print("Finished poly")

linearA_kernel_svm_clf.fit(X_train, y_train)
print("Finished linear A")
linearB_kernel_svm_clf.fit(X_train, y_train)
print("Finished linear B")
linearC_kernel_svm_clf.fit(X_train, y_train)
print("Finished linear C")
linearD_kernel_svm_clf.fit(X_train, y_train)
print("Finished linear D")
linearE_kernel_svm_clf.fit(X_train, y_train)
print("Finished linear E")


radialA_kernel_svm_clf.fit(X_train, y_train)
print("Finished radial A")
radialB_kernel_svm_clf.fit(X_train, y_train)
print("Finished radial B")
radialC_kernel_svm_clf.fit(X_train, y_train)
print("Finished radial C")
radialD_kernel_svm_clf.fit(X_train, y_train)
print("Finished radial D")
radialE_kernel_svm_clf.fit(X_train, y_train)
print("Finished radial E")
```

```python
#Predicting test and training models using polynomial kernel
predicted_polyA_test = polyA_kernel_svm_clf.predict(X_test)
predicted_polyB_test = polyB_kernel_svm_clf.predict(X_test)
predicted_polyC_test = polyC_kernel_svm_clf.predict(X_test)
predicted_polyD_test = polyD_kernel_svm_clf.predict(X_test)
predicted_polyE_test = polyE_kernel_svm_clf.predict(X_test)


predicted_polyA_train = polyA_kernel_svm_clf.predict(X_train)
predicted_polyB_train = polyB_kernel_svm_clf.predict(X_train)
predicted_polyC_train = polyC_kernel_svm_clf.predict(X_train)
predicted_polyD_train = polyD_kernel_svm_clf.predict(X_train)
predicted_polyE_train = polyE_kernel_svm_clf.predict(X_train)

#Predicting test and training models using linear kernel
predicted_linearA_test = linearA_kernel_svm_clf.predict(X_test)
predicted_linearB_test = linearB_kernel_svm_clf.predict(X_test)
predicted_linearC_test = linearC_kernel_svm_clf.predict(X_test)
predicted_linearD_test = linearD_kernel_svm_clf.predict(X_test)
predicted_linearE_test = linearE_kernel_svm_clf.predict(X_test)


predicted_linearA_train = linearA_kernel_svm_clf.predict(X_train)
predicted_linearB_train = linearB_kernel_svm_clf.predict(X_train)
predicted_linearC_train = linearC_kernel_svm_clf.predict(X_train)
predicted_linearD_train = linearD_kernel_svm_clf.predict(X_train)
predicted_linearE_train = linearE_kernel_svm_clf.predict(X_train)


#Predicting test and training models using rbf kernel
predicted_radialA_test = radialA_kernel_svm_clf.predict(X_test)
predicted_radialB_test = radialB_kernel_svm_clf.predict(X_test)
predicted_radialC_test = radialC_kernel_svm_clf.predict(X_test)
predicted_radialD_test = radialD_kernel_svm_clf.predict(X_test)
predicted_radialE_test = radialE_kernel_svm_clf.predict(X_test)

predicted_radialA_train = radialA_kernel_svm_clf.predict(X_train)
predicted_radialB_train = radialB_kernel_svm_clf.predict(X_train)
predicted_radialC_train = radialC_kernel_svm_clf.predict(X_train)
predicted_radialD_train = radialD_kernel_svm_clf.predict(X_train)
predicted_radialE_train = radialE_kernel_svm_clf.predict(X_train)
```

```python
C = [0.001, 0.01, 0.1, 1, 10]
poly_train_accur = [metrics.accuracy_score(y_true=y_test, y_pred=predicted_polyA_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_polyB_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_polyC_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_polyD_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_polyE_test)]
print("test accuracy poly C=0.001:", poly_train_accur[0], "\n")
print("test accuracy poly C=0.01:", poly_train_accur[1], "\n")
print("test accuracy poly C=0.1:", poly_train_accur[2], "\n")
print("test accuracy poly C=1:", poly_train_accur[3], "\n")
print("test accuracy poly C=10:", poly_train_accur[4], "\n")
print("------------------------------------------")

poly_test_accur = [metrics.accuracy_score(y_true=y_train, y_pred=predicted_polyA_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_polyB_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_polyC_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_polyD_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_polyE_train)]
print("train accuracy poly C=0.001:", poly_test_accur[0], "\n")
print("train accuracy poly C=0.01:", poly_test_accur[1], "\n")
print("train accuracy poly C=0.1:", poly_test_accur[2], "\n")
print("train accuracy poly C=1:", poly_test_accur[3], "\n")
print("train accuracy poly C=10:", poly_test_accur[4], "\n")

print("========================================================================")
linear_test_accur = [metrics.accuracy_score(y_true=y_test, y_pred=predicted_linearA_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_linearB_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_linearC_test), metrics.accuracy_score(y_true=y_test, y_pred=predicted_linearD_test), metrics.accuracy_score(y_true=y_test, y_pred=predicted_linearE_test)]
print("test accuracy linear C=0.001:", linear_test_accur[0], "\n")
print("test accuracy linear C=0.01:", linear_test_accur[1], "\n")
print("test accuracy linear C=0.1:", linear_test_accur[2], "\n")
print("test accuracy linear C=1:", linear_test_accur[3], "\n")
print("test accuracy linear C=10:", linear_test_accur[4], "\n")
print("------------------------------------------")

linear_train_accur = [metrics.accuracy_score(y_true=y_train, y_pred=predicted_linearA_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_linearB_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_linearC_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_linearD_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_linearE_train)]
print("train accuracy linear C=0.001:", linear_train_accur[0], "\n")
print("train accuracy linear C=0.01:", linear_train_accur[1], "\n")
print("train accuracy linear C=0.1:", linear_train_accur[2], "\n")
print("train accuracy linear C=1:", linear_train_accur[3], "\n")
print("train accuracy linear C=10:", linear_train_accur[4], "\n")

print("========================================================================")
radial_test_accur = [metrics.accuracy_score(y_true=y_test, y_pred=predicted_radialA_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_radialB_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_radialC_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_radialD_test),metrics.accuracy_score(y_true=y_test, y_pred=predicted_radialE_test)]
print("test accuracy radial C=0.001:", radial_test_accur[0], "\n")
print("test accuracy radial C=0.01:", radial_test_accur[1], "\n")
print("test accuracy radial C=0.1:", radial_test_accur[2], "\n")
print("test accuracy radial C=0.1:", radial_test_accur[3], "\n")
print("test accuracy radial C=0.1:", radial_test_accur[4], "\n")
print("------------------------------------------")

radial_train_accur = [metrics.accuracy_score(y_true=y_train, y_pred=predicted_radialA_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_radialB_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_radialC_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_radialD_train),metrics.accuracy_score(y_true=y_train, y_pred=predicted_radialE_train)]
print("train accuracy radial C=0.001:", radial_train_accur[0], "\n")
print("train accuracy radial C=0.01:", radial_train_accur[1], "\n")
print("train accuracy radial C=0.1:", radial_train_accur[2], "\n")
print("train accuracy radial C=1:", radial_train_accur[3], "\n")
print("train accuracy radial C=10:", radial_train_accur[4], "\n")


#Plotting C vs Test Accuracy
plt.plot(C, poly_test_accur, label="poly", marker="o")
plt.plot(C, linear_test_accur, label="lin", marker="o")
plt.plot(C, radial_test_accur, label="rbf", marker="o")
plt.xlabel("C")
plt.ylabel("Test Accuracy")
plt.title("C vs. Test Accuracy")
plt.legend()
plt.show()

#Plotting C vs Training Accuracy
plt.plot(C, poly_train_accur, label="poly", marker="o")
plt.plot(C, linear_train_accur, label="lin", marker="o")
plt.plot(C, radial_train_accur, label="rbf", marker="o")
plt.xlabel("C")
plt.ylabel("Training Accuracy")
plt.title("C vs. Training Accuracy")
plt.legend()
plt.show()


```
